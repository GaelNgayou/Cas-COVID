```python
import pandas as pd
from scipy.interpolate import CubicSpline
import numpy as np
```

```python
df = pd.read_csv('Covid Live.csv')
```

```python
colonnes_converties = [colonne for colonne in df.columns if colonne != ['Country,\nOther','New\nDeaths','#']]
```

```python
df[colonnes_converties] = df[colonnes_converties].replace({',': ''}, regex=True)
```

```python
nbre_nan = df.isna().sum()
```

```python
nbre_nan
```

```python
interpolation = ['Total\nDeaths', 'New\nDeaths', 'Total\nRecovered', 'Active\nCases', 'Serious,\nCritical', 'Tot Cases/\n1M pop', 'Deaths/\n1M pop', 'Total\nTests', 'Tests/\n1M pop', 'Population']
```

```python
df[interpolation] = df[interpolation].interpolate()
```
